```python
import tkinter as tk
from tkinter import messagebox

def merge_sort(arr):
    subarray_size = 1
    while subarray_size < len(arr):
        left = 0
        while left < len(arr) - 1:
            mid = min((left + subarray_size - 1), (len(arr)-1))
            right = min((left + 2 * subarray_size - 1), (len(arr)-1))
            merge(arr, left, mid, right)
            left += 2 * subarray_size
            # Display the current state of the array in the process_text widget
            process_text.insert(tk.END, f"Step: {arr}\n")
        subarray_size *= 2

def merge(arr, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid

    L = [0] * n1
    R = [0] * n2

    for i in range(n1):
        L[i] = arr[left + i]

    for j in range(n2):
        R[j] = arr[mid + 1 + j]

    i = j = 0
    k = left

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def sort_button_clicked():
    try:
        arr_str = entry.get()
        arr = list(map(int, arr_str.split()))

        # Clear the existing content in the Text widgets
        result_text.delete(1.0, tk.END)
        process_text.delete(1.0, tk.END)

        # Display the original array in the process_text widget
        process_text.insert(tk.END, f"Original Array: {arr}\n")

        merge_sort(arr)

        # Display the sorted array in the result_text widget
        result_text.insert(tk.END, str(arr))
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter a valid list of integers.")

window = tk.Tk()
window.title("Merge Sort Algorithm")
window.state('zoomed')  # Maximize the window by default

# Set a background color
window.configure(bg='#2C3E50')

# Increase font size for labels
label = tk.Label(window, text="Enter space-separated elements of the array:", font=('Arial', 18), fg='white', bg='#2C3E50')
label.pack(pady=10)

# Increase font size and height for entry widget
entry = tk.Entry(window, width=40, font=('Arial', 14))
entry.pack(pady=10)

# Increase font size and height for button, change color and add border
sort_button = tk.Button(window, text="Sort", command=sort_button_clicked, font=('Arial', 14), height=2, bg='#3498DB', fg='white', bd=4)
sort_button.pack(pady=20)

# Add another Text widget for displaying the process steps, change font color
process_label = tk.Label(window, text="Process Steps:", font=('Arial', 18), fg='white', bg='#2C3E50')
process_label.pack(pady=5)

process_text = tk.Text(window, width=40, height=15, font=('Arial', 14), bg='#34495E', fg='white')
process_text.pack(pady=10)  # Placed under the button

# Change label to "Result:", change font color
result_label = tk.Label(window, text="Result:", font=('Arial', 18), fg='white', bg='#2C3E50')
result_label.pack(pady=5)

# Create a Text widget for displaying the result, change font color
result_text = tk.Text(window, width=40, height=5, font=('Arial', 14), bg='#34495E', fg='white')
result_text.pack(pady=10)  # Placed under the process_text

window.mainloop()
```
